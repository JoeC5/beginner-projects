```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\v-joecamp\OneDrive - Microsoft\Desktop\Jupyter Notebook\company_financials.csv')
df.head()
```

```python
df_copy = df.copy()
df_copy.info()
```

```python
#converting the date column to datetime object
df_copy['date'] = pd.to_datetime(df_copy['date'])
```

```python
#adding a week column to the Dataframe
df_copy['number_week'] = df_copy['date'].dt.isocalendar().week
df_copy
```

```python
week_sales_1 = df_copy[df_copy['number_week'] ==6]['sales'].sum()
week_sales_2 = df_copy[df_copy['number_week'] ==7]['sales'].sum()
week_sales_3 = df_copy[df_copy['number_week'] ==8]['sales'].sum()

print(f"Week 1 sales: ${week_sales_1}")
print(f"Week 2 sales: ${week_sales_2}")
print(f"Week 3 sales: ${week_sales_3}")
      
```

```python
plt.bar(["week1", "week2", "week3"], [week_sales_1, week_sales_2, week_sales_3])
plt.xlabel('Weeks')
plt.ylabel('Sales ($)')
plt.title('Week on Week Sales Performance')
plt.show()
```

```python
#calculate total costs
total_costs = df_copy['direct_costs'] + df_copy['overheads']
#calculate the percentage of direct costs to toal costs
total_cost_pct = ((df_copy['direct_costs']/total_costs)* 100).mean()

print(f"The percent of direct costs to total costs: {total_cost_pct:.2f}%")
```

```python
#calculating the cost variation
cost_variation = (total_costs.max() - total_costs.min())/total_costs.mean()*100
print(f"The mean cost variation: {cost_variation:.2f}%")
```

```python
#adding a profit column
df_copy['profit'] = df_copy['sales'] - df_copy['direct_costs'] - df_copy['overheads']
#calculating profits for Mondays
monday_profit = df_copy[df_copy["day_of_week"] =="Monday"]["profit"].sum()
#calculting profits for Fridays
friday_profit = df_copy[df_copy["day_of_week"] == "Friday"]["profit"].sum()

difference_profit = friday_profit - monday_profit
print(f"The difference in profit between Fridys and Mondays is ${difference_profit}")
```

```python
plt.bar(["Modays", "Fridays"], [monday_profit, friday_profit], color=['blue', 'green'])
plt.xlabel('Days')
plt.ylabel('Profit($)')
plt.title("Mondays vs Friday Profit Performance")
plt.show()
```

```python
#visualizing profit trend over time
df_copy.set_index('date')['profit'].plot(figsize = (8, 6))
plt.title('Daiy Profitability')
plt.xlabel('Date')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df_copy[['sales', 'direct_costs', 'overheads']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, fmt=".1f")
plt.title("Sales and Costs Correlation")
plt.show()
```
